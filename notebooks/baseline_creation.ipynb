# Baseline creation

```python
from pathlib import Path
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

from shapely.geometry import Point, Polygon
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import GroupKFold, cross_val_score
from sklearn.preprocessing import normalize

RANDOM_STATE = 42
EMBEDDING_DIM = 64
N_PROTOTYPES = 16
PCA_TARGET_DIM = 128
USE_SYNTHETIC_DATA = True
np.random.seed(RANDOM_STATE)
```

```python
# helper functions

def l2_normalize_rows(X: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    """L2 normalize the rows of X."""
    X = np.asarray(X, dtype=np.float32)
    norms = np.linalg.norm(X, axis=1, keepdims=True)
    norms = np.maximum(norms, eps)
    return X / norms

def ensure_embedding_column(
    df: pd.DataFrame,
    embedding_col: str = 'embedding',
    prefix: str = 'e_'
) -> pd.DataFrame:
    
    if embedding_col in df.columns:
        out = df.copy()
        out[embedding_col] = out[embedding_col].apply(lambda x: np.asarray(x,dtype=np.float32))
        return out
    
    emb_cols = [c for c in df.columns if c.startswith(prefix)]

    if not emb_cols:
        raise ValueError(
            f'Could not find `{embedding_col}` or any columns starting with `{prefix}`.'
        )
    
    def sort_key(c):
        suffix = c[len[prefix]]
        return int(suffix) if suffix.isdigit() else suffix
    
    emb_cols = sorted(emb_cols,key=sort_key)
    out = df.copy()
    out[embedding_col] = list(out[emb_cols].to_numpy(dtype=np.float32))

    return out

def hull_vertices_to_polygons(
    hull_vertices : pd.DataFrame,
    port_col:str = 'port_id',
    part_col:str = 'part_id',
    order_col:str = 'vertex_order',
    lat_col:str = 'lat',
    lon_col:str = 'lon',
    crs:str = 'EPSG:4326'
) -> gpd.GeoDataFrame:
    records = []
    required = [port_col,part_col,order_col,lat_col,lon_col]
    missing = [c for c in required if c not in hull_vertices.columns]
    
    if missing:
        raise ValueError(f'Missing required hull columns: {missing}')

    hull_vertices = hull_vertices.sort_values([port_col,part_col,order_col]).copy()

    for (port_id, part_id), group in hull_vertices.groupby([port_col, part_col],sort=False):
        coords = list(zip(group[lon_col].to_numpy(),group[lat_col].to_numpy()))
        if len(coords) < 3:
            continue
        poly = Polygon(coords)
        if not poly.is_valid:
            poly = poly.buffer(0)
        
        records.append(
            {
                port_col:port_id,
                part_col:part_id,
                'geometry':poly,
            }
        )
    
    parts_gdf = gpd.GeoDataFrame(records,geometry='geometry',crs=crs)
    if len(parts_gdf) == 0:
        raise ValueError('No valid polygons found.')
    return parts_gdf

def assign_patches_to_parts(
    patches_df: pd.DataFrame,
    parts_gdf: gpd.GeoDataFrame,
    lat_col: str = 'lat',
    lon_col: str = 'lon',
    crs: str = 'EPSG:4326'
) -> gpd.GeoDataFrame:
    patches_gdf = gpd.GeoDataFrame(
        patches_df.copy(),
        geometry=gpd.points_from_xy(patches_df[lon_col],patches_df[lat_col]),
        crs=crs,
    )
    joined = gpd.sjoin(
        patches_gdf,
        parts_gdf[[c for c in parts_gdf.columns if c!='geometry'] + ['geometry']],
        how='inner',
        predicate='within',
    ).drop(columns='index_right')
    return joined

def descriptor_feature_names(embedding_dim: int, n_prototypes: int) -> list[str]:
    return (
        [f'mean_{i:03d}' for i in range(embedding_dim)] 
        + ['concentration']
        + [f'std_{i:03d}' for i in range(embedding_dim)]
        + [f'proto_{i:03d}' for i in range(n_prototypes)]
    )

def patch_group_descriptor(
    X: np.ndarray,
    prototype_centers: np.ndarray,
    weights: np.ndarray | None = None,
) -> np.ndarray:
    X = l2_normalize_rows(np.asarray(X,dtype=np.float32))
    n,d = X.shape
    if weights is None:
        w = np.ones(n, dtype=np.float32)
    else:
        np.asarray(weights,dtype=np.float32).reshape(-1)
        if len(w)!=n:
            raise ValueError("Weights must have the same length as X.")
    w = w/np.maximum(w.sum(),1e-12)

    mean_vec = (w[:,None] * X).sum(axis=0)
    concentration = float(np.linalg.norm(mean_vec))
    mean_direction = mean_vec / max(concentration,1e-12)

    diag_std = np.sqrt((w[:,None] * (X - mean_vec)**2).sum(axis=0))
    sims = X @ prototype_centers.T
    assign = sims.argmax(axis=1)
    hist = np.bincount(assign, weights=w, minlength=len(prototype_centers)).astype(np.float32)

    return np.concatenate([mean_direction,[concentration],diag_std,hist]).astype(np.float32)

def build_group_descriptors(
    df: pd.DataFrame,
    group_cols: list[str],
    prototype_centers: np.ndarray,
    embedding_col: str='embedding',
    patch_weight_col: str | None = None,      
) -> tuple[pd.DataFrame, list[str]]:
    rows = []
    d = len(np.asarray(df.iloc[0][embedding_col]))
    k = len(prototype_centers)
    feature_cols = descriptor_feature_names(d,k)

    for group_key,group in df.groupby(group_cols, sort=False):
        if not isinstance(group_key,tuple):
            group_key = (group_key,)
        X = np.vstack(group[embedding_col].to_numpy())
        weights = None

        if patch_weight_col is not None and patch_weight_col in group.columns:
            weights = group[patch_weight_col].to_numpy(dtype=np.float32)
        
        desc = patch_group_descriptor(X, prototype_centers=prototype_centers,weights=weights)

        row = {col:val for col,val in zip(group_cols,group_key)}
        row['n_patches'] = len(group)
        row.update({col : float(val) for col, val in zip(feature_cols, desc)})
        rows.append(row)
    
    return pd.DataFrame(rows),feature_cols

def aggregate_part_descriptors_to_ports(
    part_desc: pd.DataFrame,
    port_group_cols: list[str],
    feature_cols: list[str],
    weight_col: str='n_patches',  
) -> tuple[pd.DataFrame, list[str]]:
    rows = []
    out_feature_cols = (
        [f'part_mean__{c}' for c in feature_cols]
        + [f'part_std__{c}' for c in feature_cols]
    )

    for group_key, group in part_desc.groupby(port_group_cols,sort=False):
        if not isinstance(group_key,tuple):
            group_key = (group_key,)
        
        F = group[feature_cols].to_numpy(dtype=np.float32)
        w = group[weight_col].to_numpy(dtype=np.float32)
        w = w/np.maximum(w.sum(),1e-12)

        mean_feat = (w[:,None] * F).sum(axis=0)
        std_feat = np.sqrt((w[:,None] * (F - mean_feat)**2).sum(axis=0))

        row = {col : val for col, val in zip(port_group_cols, group_key)}
        row['n_parts'] = int(len(group))
        row['total_part_weight'] = float(group[weight_col].sum())
        row.update({col : float(val) for col,val in zip(out_feature_cols,np.concatenate([mean_feat,std_feat]))})
        rows.append(row)
    
    return pd.DataFrame(rows), out_feature_cols

def compress_port_features(
    df: pd.DataFrame,
    feature_cols: list[str],
    target_dim: int=128,
    random_state=42,
) -> tuple[pd.DataFrame, list[str], PCA | None]:
    n_rows = len(df)
    n_components = min(target_dim, len(feature_cols),max(1,n_rows-1))

    out = df.copy()

    if n_components < 2:
        raw = out[feature_cols].to_numpy(dtype=np.float32)
        raw = l2_normalize_rows(raw)
        emb_cols = [f'port_emb_{i:03d}' for i in range(raw.shape[1])]
        out[emb_cols] = raw
        return out, emb_cols, None
    
    pca = PCA(n_components=n_components, random_state=random_state)
    Z = pca.fit_transform(out[feature_cols].to_numpy(dtype=np.float32))
    Z = l2_normalize_rows(Z)

    emb_cols = [f'port_emb_{i:03d}' for i in range(Z.shape[1])]
    out[emb_cols] = Z
    return out, emb_cols, pca

def recall_at_k_same_port(
    port_df: pd.DataFrame,
    embedding_cols: list[str],
    port_id_col:str = 'port_id',
    view_col: str='year',
    k: int = 1
) -> float:
    X = port_df[embedding_cols].to_numpy(dtype=np.float32)
    X = l2_normalize_rows(X)
    sims = cosine_similarity(X)
    np.fill_diagonal(sims, -np.inf)
    hits = []

    for i in range(len(port_df)):
        nn = np.argsort(-sims[i])[k:]
        same_port_other_view = any(
            (port_df.iloc[j][port_id_col] == port_df.iloc[i][port_id_col])
            and (port_df.iloc[j][view_col] == port_df.iloc[i][view_col])
            for j in nn
        )
        hits.append(same_port_other_view)

    return float(np.mean(hits)) 
```

```python
def random_unit_vectors(n: int, d: int, seed: int = 42) -> np.ndarray:
    """Generate n random unit vectors in d dimensions."""
    rng = np.random.default_rng(seed)
    X = rng.normal(size=(n, d)).astype(np.float32)
    return l2_normalize_rows(X)

def rectangle_vertices(center_lon, center_lat, width, height):
    half_w = width / 2.0
    half_h = height / 2.0
    return [
        (center_lon - half_w, center_lat - half_h),  # Bottom-left
        (center_lon + half_w, center_lat - half_h),  # Bottom-right
        (center_lon + half_w, center_lat + half_h),  # Top-right
        (center_lon - half_w, center_lat + half_h)   # Top-left
    ]

def sample_points_in_rectangle(center_lon, center_lat, width, height, n, rng):
    lons = rng.uniform(center_lon - width/2.0, center_lon + width/2.0,size=n)
    lats = rng.uniform(center_lat - height/2.0, center_lat + height/2.0,size=n)
    return lons, lats

def sample_embeddings_from_mixture(centers, probs, n, noise_scale, rng):
    probs = np.asarray(probs, dtype=np.float32)
    probs = probs/probs.sum()
    assignments = rng.choice(len(probs),size=n,p=probs)
    X = centers[assignments] + rng.normal(scale=noise_scale,size=(n,centers.shape[1]))
    return l2_normalize_rows(X.astype(np.float32)), assignments

def make_synthetic_port_data(embedding_dim=64, seed=42):
    rng = np.random.default_rng(seed)

    semantic_names=[
        'water',
        'container',
        'tank',
        'warehouse',
        'road',
        'urban_edge'
    ]

    semantic_centers = random_unit_vectors(len(semantic_names),embedding_dim,seed=seed)

    port_templates = [
        {
            "port_id" : "port_A",
            "base_lat" : 1.250,
            "base_lon" : 103.00,
            "parts" : [
                {
                    "part_id" : "A_1",
                    "dx" : -0.020,
                    "dy" : 0.000,
                    "w" : 0.030,
                    "h" : 0.015,
                    "mix" : [0.12,0.18,0.00,0.40,0.18,0.12]
                },
                {
                    "part_id" : "A_2",
                    "dx" : 0.020,
                    "dy" : 0.000,
                    "w" : 0.028,
                    "h" : 0.018,
                    "mix" : [0.12,0.18,0.00,0.40,0.18,0.12]
                }
            ],
        },
        {
            "port_id" : "port_B",
            "base_lat" : 25.280,
            "base_lon" : 55.280,
            "parts" : [
                {
                    "part_id" : "B_1",
                    "dx" : -0.018,
                    "dy" : 0.002,
                    "w" : 0.026,
                    "h" : 0.016,
                    "mix" : [0.18,0.05,0.48,0.10,0.12,0.07]
                },
                {
                    "part_id" : "B_2",
                    "dx" : 0.018,
                    "dy" : -0.003,
                    "w" : 0.030,
                    "h" : 0.018,
                    "mix" : [0.28,0.05,0.35,0.07,0.15,0.10]
                },
            ],
        },
        {
            "port_id" : "port_C",
            "base_lat" : -23.960,
            "base_lon" : -46.300,
            "parts" : [
                {
                    "part_id" : "C_1",
                    "dx" : -0.020,
                    "dy" : 0.000,
                    "w" : 0.032,
                    "h" : 0.016,
                    "mix" : [0.16,0.08,0.14,0.22,0.18,0.22]
                },
                {
                    "part_id" : "C_2",
                    "dx" : 0.018,
                    "dy" : -0.003,
                    "w" : 0.030,
                    "h" : 0.018,
                    "mix" : [0.14,0.14,0.12,0.18,0.18,0.24]
                },
            ],
        },
    ]
    years = [2024,2025]
    hull_rows = []
    patch_rows = []

    for port in port_templates:
        for part in port['parts']:
            center_lon = port['base_lon'] + part['dx']
            center_lat = port['base_lat'] + part['dy']

            vertices = rectangle_vertices(center_lon,center_lat,part['w'],part['h'])

            for order, (lon, lat) in enumerate(vertices):
                hull_rows.append({
                    'port_id':port['port_id'],
                    'part_id':part['part_id'],
                    'vertex_order':order,
                    'lat':lat,
                    'lon':lon,
                })

            for year in years:
                n_patches = 180
                lons, lats = sample_points_in_rectangle(
                    center_lon=center_lon,
                    center_lat=center_lat,
                    width=part['w'] * 0.90,
                    height=part['h'] * 0.90,
                    n = n_patches,
                    rng=rng
                )

                mix = np.array(part['mix'],dtype=np.float32)
                mix = np.clip(mix + rng.normal(scale=0.008,size=len(mix)),0.001,None)
                mix = mix / mix.sum()

                X, assignments= sample_embeddings_from_mixture(
                    centers = semantic_centers,
                    probs=mix,
                    n=n_patches,
                    noise_scale=0.07,
                    rng=rng
                )
                for i in range(n_patches):
                    patch_rows.append(
                        {
                            'patch_id':f"{port['port_id']}_{part['part_id']}_{year}_{i:04d}",
                            'lat':float(lats[i]),
                            'lon':float(lons[i]),
                            'year':int(year),
                            'embedding':X[i],
                            'true_semantic':semantic_names[int(assignments[i])],
                        }
                    )
    hull_df = pd.DataFrame(hull_rows)
    patches_df = pd.DataFrame(patch_rows)
    return patches_df, hull_df, semantic_names

synthetic_patches_df, synthetic_hulls_df, synthetic_semantic_names = make_synthetic_port_data(
    embedding_dim=EMBEDDING_DIM,
    seed = RANDOM_STATE,
)

synthetic_patches_df.head(), synthetic_hulls_df.head()
```

```python
if USE_SYNTHETIC_DATA:
    patches_df = synthetic_patches_df.copy()
    hull_vertices_df = synthetic_hulls_df.copy()
```

```python
patches_df = ensure_embedding_column(patches_df,embedding_col='embedding',prefix='e_')
print('patches_df shape:', patches_df.shape)
print('hull_vertices_df shape', hull_vertices_df.shape)
```

# Build port-part Polygons from hull vertices

```python
parts_gdf = hull_vertices_to_polygons(
    hull_vertices_df,
    port_col='port_id',
    part_col='part_id',
    order_col='vertex_order',
    lat_col='lat',
    lon_col='lon'
)
```

```python
parts_area = parts_gdf.to_crs('EPSG:6933').copy()
parts_gdf['part_area_m2'] = parts_area.area.values

parts_gdf.head()
```

# Spatially join patch centroids to parts

```python
joined = assign_patches_to_parts(
    patches_df,
    parts_gdf,
    lat_col='lat',
    lon_col='lon'
)

keep_cols = [c for c in joined.columns if c in {
    'patch_id','lat','lon','year','embedding','true_semantic',
    'port_id','part_id','part_area_m2','geometry'
}]
joined = joined[keep_cols].copy()

print("Joined patch rows:",len(joined))
print('Unique ports:',joined['port_id'].nunique())
print('Unique parts:',joined['part_id'].nunique())
joined.head()
```

# Fit a global prototype codebook on patch embeddings

```python
X_all = np.vstack(joined['embedding'].to_numpy())
X_all = l2_normalize_rows(X_all)

kmeans = MiniBatchKMeans(
    n_clusters= N_PROTOTYPES,
    random_state=RANDOM_STATE,
    batch_size=1024,
    n_init=10
)

kmeans.fit(X_all)
prototype_centers = l2_normalize_rows(kmeans.cluster_centers_)
```

```python
prototype_centers.shape
```

# Build one descriptor per port part

```python
part_desc, part_feature_cols = build_group_descriptors(
    joined,
    group_cols= ['port_id','year','part_id'],
    prototype_centers=prototype_centers,
    embedding_col='embedding',
    patch_weight_col=None
)

part_desc = part_desc.merge(
    parts_gdf[['port_id','part_id','part_area_m2']],
    on=['port_id','part_id'],
    how='left'
)

print('part_desc shape', part_desc.shape)
part_desc[['port_id','year','part_id','n_patches','part_area_m2']].head()
```

```python
proto_cols = [c for c in part_feature_cols if c.startswith('proto_')]
display_cols = ['port_id','year','part_id'] + proto_cols[:8]
part_desc[display_cols].head(6)
```

# Aggregate part descriptors 

```python
WEIGHT_COL = 'n_patches'

port_desc, port_feature_cols = aggregate_part_descriptors_to_ports(
    part_desc,
    port_group_cols=['port_id','year'],
    feature_cols=part_feature_cols,
    weight_col=WEIGHT_COL
)

print("port_desc shape",port_desc.shape)
port_desc[['port_id','year','n_parts','total_part_weight']].head()
```

```python
port_desc
```

# Compress to final port embedding

```python
port_desc_compact, port_embedding_cols, pca_model = compress_port_features(
    port_desc,
    feature_cols = port_feature_cols,
    target_dim = PCA_TARGET_DIM,
    random_state=RANDOM_STATE
)

print('Final port embedding dimension:', len(port_embedding_cols))
port_desc_compact[['port_id','year']+port_embedding_cols[:5]].head()
```

# Retrieval comparison (validation)

```python
r_at_1 = recall_at_k_same_port(
    port_desc_compact,
    embedding_cols=port_embedding_cols,
    port_id_col='port_id',
    view_col='year',
    k = 1
)

r_at_2 = recall_at_k_same_port(
    port_desc_compact,
    embedding_cols=port_embedding_cols,
    port_id_col='port_id',
    view_col='year',
    k = 2
)

print(f'Recall@1 (same port, other-year retrieval): {r_at_1:.3f}')
print(f'Recall@2 (same port, other-year retrieval): {r_at_1:.3f}')
```

```python
X_port = l2_normalize_rows(port_desc_compact[port_embedding_cols].to_numpy(dtype=np.float32))
sims = cosine_similarity(X_port)
np.fill_diagonal(sims, -np.inf)

neighbour_rows = []

for i in range(len(port_desc_compact)):
    best_j = int(np.argmax(sims[i]))
    neighbour_rows.append(
        {
            'query_port': port_desc_compact.iloc[i]['port_id'],
            'query_year': int(port_desc_compact.iloc[i]['year']),
            'nn_port': port_desc_compact.iloc[best_j]['port_id'],
            'nn_year': int(port_desc_compact.iloc[best_j]['year']),
            'cosine_similarity':float(sims[i,best_j])
        }
    )

pd.DataFrame(neighbour_rows)
```
